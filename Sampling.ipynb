```python
import random
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
from scipy.spatial.distance import cosine, euclidean, minkowski, chebyshev, cityblock

import pickle as pkl
```

```python
# np.random.seed(seed=0)
item_num = 10000
item_vec_size = 2

## 離散：[0,5]
# item_prof_vecs = np.random.randint(0, 6, (item_num, item_vec_size))
## 連続：[0,1)
item_prof_vecs = -1 * np.random.uniform(-1, 0, (item_num, item_vec_size))
```

```python
x = 2  ## 選択item数

# np.random.seed(seed=1)
item_id_list = list(range(0, item_num))

user_prof_vecs = np.empty((0, item_vec_size))
for _ in range(0, item_num):
    ## item_idの取得
    items = np.random.choice(item_id_list, np.random.randint(1, 3))
#     items = np.random.choice(item_id_list, x)
    ## item_vecsの取得
    item_vecs = np.empty((0, item_vec_size), dtype=np.int)
    for item_id in items:
        item_vecs = np.append(item_vecs, item_prof_vecs[item_id].reshape(1, item_vec_size), axis=0)
    ## item_vecsを平均
    item_vec = np.average(item_vecs, axis=0)
    user_prof_vecs = np.append(user_prof_vecs, item_vec.reshape(1, item_vec_size), axis=0)
```

```python
item_arr = item_prof_vecs.reshape((2, item_num))
user_arr = user_prof_vecs.reshape((2, item_num))
```

```python
## item選択数 1
figsize(15, 15)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.scatter(item_arr[0], item_arr[1], c='m')
ax.scatter(user_arr[0], user_arr[1], c='c')

ax.set_xlabel('x')
ax.set_ylabel('y')

fig.show()
```
