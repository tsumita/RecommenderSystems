# データセット作成
## 手順
1. itemのプロファイルを定義（itemプロファイルベクトル$I_{P_j}$をitem数だけ作成）
2. userの嗜好itemを選択
3. userのプロファイルを定義（userプロファイルベクトル$U_{P_i}$をuser数だけ作成）
4. $U_{P_i}とI_{P_j}$からuser $i$のitem $j$に対する評価値を決定。全user全itemに対して求め、matrixを作成する
5. 評価値の欠損率を変えて欠損データを複数パターン作成

### 2. userの嗜好itemを選択
- user数：10,000  
- 関心のあるitemを$x$個選択してもらうことを想定
- ランダムにitemを$x$個選択


### 3. userのプロファイルを定義  
- 関心のあるitemのプロファイルを平均したものをuserのプロファイルとする  

---

```python
# x = 1  ## 選択item数
user_num = 10000

np.random.seed(seed=1)
item_id_list = list(range(0, user_num))

user_prof_vecs = np.empty((0, item_vec_size))
for _ in range(0, user_num):
    ## item_idの取得
    items = np.random.choice(item_id_list, np.random.randint(2,6))
#     items = np.random.choice(item_id_list, x)
    ## item_vecsの取得
    item_vecs = np.empty((0, item_vec_size), dtype=np.int)
    for item_id in items:
        item_vecs = np.append(item_vecs, item_prof_vecs[item_id].reshape(1, item_vec_size), axis=0)
    ## item_vecsを平均
    item_vec = np.average(item_vecs, axis=0)
    user_prof_vecs = np.append(user_prof_vecs, item_vec.reshape(1, item_vec_size), axis=0)
```

### 4. $U_{P_i}とI_{P_j}$からuser $i$のitem $j$に対する評価値を決定。matrixを作成する  
- ベクトル$U_{P_i}, I_{P_j}$の$cos$類似度を求める  
- $cos$類似度に基づく乱数でサンプリング。user $i$のitem $j$に対する評価値を算出する  
- 一旦、userごとのitem評価ベクトルで$Max-Min$正規化  

---

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
#     return cosine(v1, v2)

def euclid_sim(v1, v2):
    return 1 / (1 + euclidean(v1, v2))

def manhattan_sim(v1, v2):
    return 1 / (1 + cityblock(v1, v2))

def chebyshev_sim(v1, v2):
    return 1 / (1 + chebyshev(v1, v2))

def cos_euclid_sim(v1, v2):
    return (cos_sim(v1, v2) + euclid_sim(v1, v2)) / 2


## item, userベクトルから評価値計算
def cal_eval_val(v1, v2, scale=0.05):
    ## 類似度計算
    similarity = cos_sim(v1, v2)
#     similarity = euclid_sim(v1, v2)
#     similarity = manhattan_sim(v1, v2)
#     similarity = chebyshev_sim(v1, v2)
#     similarity = cos_euclid_sim(v1, v2)
#     return similarity
    
    ## 平均→評価値, 分散→0.1の正規分布でサンプリング
    eval_val = np.random.normal(similarity, scale)
#     if eval_val > 1:
#         return 1.
#     elif eval_val < 0:
#         return 0.
    return eval_val
```

```python
matrix = np.zeros(shape=(user_num,item_num))
for u_id, user_vec in enumerate(user_prof_vecs):
    matrix[u_id] = np.array(list(map(lambda item_vec:cal_eval_val(user_vec, item_vec) , item_prof_vecs)))
    
# matrix = (matrix - matrix.min()) / (matrix.max() - matrix.min())
matrix = (matrix - matrix.min(axis=1).reshape((item_num,1))) / (matrix.max(axis=1) - matrix.min(axis=1)).reshape((item_num,1))
df_matrix = pd.DataFrame(matrix, index=['user_'+str(i) for i in range(item_num)], columns=['item_'+str(i) for i in range(item_num)])
```

```python
value_arr = df_matrix.values.flatten()
mean = np.mean(value_arr)
variance = np.var(value_arr)
print("Mean: {}, Variance:{}".format(mean, variance))
figsize(15, 6)
plt.hist(value_arr, bins=70, color="y", normed=True, histtype = "stepfilled")
plt.title("cos sim")
plt.xlabel("valid value")
plt.ylabel("frequency")
```

### 5. 評価値の欠損率を変えて欠損データを複数パターン作成
- 欠損率$10\%$刻み  

---

```python
df_matrix = pd.read_csv("../dataset/origin_matrix.csv", index_col=0)
```

```python
threshold = math.floor(len(df_matrix)/5*4)
df_mtrx_trn = df_matrix[:threshold]
df_mtrx_tst = df_matrix[threshold:]
```

#### train set

```python
def random_del(df, del_rate):
    df_ = df.copy()
    random.seed(7)
    arr_ = df_.values
    rows, lines = df_.shape
    delete_size = math.floor(del_rate*rows*lines)
    target_id_arr = np.random.choice(range(rows*lines), delete_size, replace=False)
    target_uid_arr = target_id_arr // lines
    target_iid_arr = target_id_arr % lines
    
    train_id_arr = np.delete(np.array(range(rows*lines)), target_id_arr)
    train_uid_arr = train_id_arr // lines
    train_iid_arr = train_id_arr % lines
    
    train_dataset = [0]*len(train_id_arr)
    tst_data_idx = 0
    for uid, iid in zip(train_uid_arr, train_iid_arr):
        train_dataset[tst_data_idx] = [uid, iid, arr_[uid][iid]]
        tst_data_idx += 1
    
    for uid, iid in zip(target_uid_arr, target_iid_arr):
        arr_[uid][iid] = None
        
    df_ = pd.DataFrame(arr_, index=['user_'+str(i) for i in range(rows)], columns=['item_'+str(i) for i in range(lines)])
    
    return df_, train_dataset
```

```python
del_rate = 0.1
df_trn, train_dataset = random_del(df_mtrx_trn, del_rate)
train_dataset = pd.DataFrame(train_dataset)
```

#### test set

```python
def random_del_tst(df, del_rate):
    df_ = df.copy()
    np.random.seed(seed=8)
    arr_ = df_.values
    rows, lines = df_.shape
    delete_size = math.floor(del_rate*rows*lines)
    target_id_arr = np.random.choice(range(rows*lines), delete_size, replace=False)
    target_uid_arr = target_id_arr // lines
    target_iid_arr = target_id_arr % lines
    
    test_dataset = [0]*delete_size
    tst_data_idx = 0
    for uid, iid in zip(target_uid_arr, target_iid_arr):
        test_dataset[tst_data_idx] = [uid, iid, arr_[uid][iid]]
        arr_[uid][iid] = None
        tst_data_idx += 1
        
    df_ = pd.DataFrame(arr_, index=['user_'+str(i) for i in range(rows)], columns=['item_'+str(i) for i in range(lines)])
    
    return df_, test_dataset
```

```python
del_rate = 0.1
df_tst, test_dataset = random_del_tst(df_mtrx_tst, del_rate)
test_dataset = pd.DataFrame(test_dataset)
```

```python
del_rate = 0.1
df_trn, train_dataset = random_del(df_mtrx_trn, del_rate)
df_tst, test_dataset_all = random_del_tst(df_mtrx_tst, del_rate)

dir_name = "del_{}per".format(int(del_rate*100))
try:
    os.makedirs("../dataset/{}".format(dir_name))
except FileExistsError:
    pass

train_dataset = pd.DataFrame(train_dataset)
test_dataset_all = pd.DataFrame(test_dataset_all)

df_trn.to_csv("../dataset/{}/matrix_trn.csv".format(dir_name))
df_tst.to_csv("../dataset/{}/matrix_tst.csv".format(dir_name))
train_dataset.to_csv("../dataset/{}/train_data.csv".format(dir_name), header=False, index=False)
test_dataset.to_csv("../dataset/{}/test_data.csv".format(dir_name), header=False, index=False)
test_dataset_all.to_csv("../dataset/{}/test_dataset_all.csv".format(dir_name), header=False, index=False)
```
