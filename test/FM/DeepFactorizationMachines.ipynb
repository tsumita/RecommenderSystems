```python
def make_valdata(val_case, test_size=20000, select_cols=False):
    df_train = pd.read_csv("../../dataset/ml-100k/valcase_{}/train_data_with_datetime.csv".format(val_case), names=['user_id','movie_id','rating','time_stamp'])
    df_test  = pd.read_csv("../../dataset/ml-100k/valcase_{}/test_data_with_datetime.csv".format(val_case),  names=['user_id','movie_id','rating','time_stamp'])
    
    df_train = pd.merge(df_train, df_user, how='left', on='user_id')
    df_train = pd.merge(df_train, df_item, how='left', on='movie_id')
    df_test = pd.merge(df_test, df_user, how='left', on='user_id')
    df_test = pd.merge(df_test, df_item, how='left', on='movie_id')
    
    df_train = df_train.rename(columns={"Children's": 'Children'})
    df_test  = df_test.rename(columns={"Children's": 'Children'})
    
    genre_list = ['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
    df_train['genre'] = ''
    df_test['genre']  = ''
    for genre in genre_list:
        df_train['genre'] = df_train['genre'] + df_train[genre].apply(lambda x: genre if x == 1 else '')+'|'
        df_test['genre']  = df_test['genre']  + df_test[genre].apply(lambda x: genre if x == 1 else '')+'|'
        
#     df_train = df_train.drop(columns=genre_list)
#     df_test  = df_test.drop(columns=genre_list)
    
    df_train['genre'] = df_train['genre'].apply(lambda x: re.sub('\|+', '|', x).lstrip('|').rstrip('|'))
    df_test['genre']  = df_test['genre'].apply(lambda x: re.sub('\|+', '|', x).lstrip('|').rstrip('|'))
    
    return df_train, df_test
```

```python
def plot_result(mae_trn, mae, rmse_trn, rmse, mse_trn, mse, colors=['#dc143c','#8b0000','#ff00ff','#ff8c00','#ffff00','#adff2f','#008000','#808000','#8b008b','#4169e1','#5f9ea0','#00ced1','#696969'], steps=-5):
    mae_list  = [v for v in mae]
    rmse_list = [v for v in rmse]
    mse_list  = [v for v in mse]
    df_result = pd.DataFrame([mae_list, rmse_list, mse_list], index=['MAE','RMSE','MSE'])
    horizon = [i for i in range(1,len(mae)+1)]

    ## MAE
    plt.figure(figsize=(10,6))
    plt.plot(horizon, mae_trn, color=colors[0], label='train_data')
    plt.plot(horizon, mae, color=colors[7], label='test_data')
    plt.title("Mean Absolute Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("MAE", fontsize=12)
#     plt.legend(loc='upper right', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()
    
    ## RMSE
    plt.figure(figsize=(10,6))
    plt.plot(horizon, rmse_trn, color=colors[0], label='train_data')
    plt.plot(horizon, rmse, color=colors[7], label='test_data')
    plt.title("Root Mean Square Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("RMSE", fontsize=12)
#     plt.legend(loc='upper right', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()
    
    ## MSE
    plt.figure(figsize=(10,6))
    plt.plot(horizon, mse_trn, color=colors[0], label='train_data')
    plt.plot(horizon, mse, color=colors[7], label='test_data')
    plt.title("Mean Square Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("MSE", fontsize=12)
#     plt.legend(loc='upper right', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()
    
#     return df_result
    return pd.DataFrame(df_result.loc[:, df_result.loc['MAE'].idxmin()])
```

```python
key2index = {}

def split(x):
    key_ans = x.split('|')
    for key in key_ans:
        if key not in key2index:
            # Notice : input value 0 is a special "padding",so we do not use 0 to encode valid feature for sequence input
            key2index[key] = len(key2index) + 1
    return list(map(lambda x: key2index[x], key_ans))

def valid_deepfm(df_train, df_test, dfm_params, sparse_features, dense_features, sequence_features, params):
    
    n_iter = params['n_iter']
    
    df_all = pd.concat([df_train.copy(), df_test.copy()])
    
    df_train_trans = df_train.copy()
    df_test_trans  = df_test.copy()
    
    mae, rmse, mse, mae_trn, rmse_trn, mse_trn = [0 for _ in range(n_iter)], [0 for _ in range(n_iter)],[0 for _ in range(n_iter)],[0 for _ in range(n_iter)],[0 for _ in range(n_iter)],[0 for _ in range(n_iter)]
    
    target = ['rating']
    
    # 1.Label Encoding for sparse features,and do simple Transformation for dense features
    for feat in sparse_features:
        lbe = LabelEncoder()
        df_train_trans[feat] = lbe.fit_transform(df_train_trans[feat])
        df_test_trans[feat]  = lbe.fit_transform(df_test_trans[feat])
        
        
    seq_feat_list = []
    if len(sequence_features) > 0:
        # preprocess the sequence feature
        global key2index
        key2index = {}
        genres_list = list(map(split, df_all['genre'].values))
        genres_length = np.array(list(map(len, genres_list)))
        max_len = max(genres_length)

        genres_list_trn = list(map(split, df_train['genre'].values))
        # Notice : padding=`post`
        genres_list_trn = pad_sequences(genres_list_trn, maxlen=max_len, padding='post',)

        genres_list_tst = list(map(split, df_all['genre'].values))
        # Notice : padding=`post`
        genres_list_tst = pad_sequences(genres_list_tst, maxlen=max_len, padding='post',)
        
        # Notice : value 0 is for padding for sequence input feature
        seq_feat_list    = [VarLenFeat('genre', len(key2index) + 1, max_len, 'mean')]

        
    # 2.count #unique features for each sparse field
    sparse_feat_list = [SingleFeat(feat,df_all[feat].nunique()) for feat in sparse_features]
    dense_feat_list  = [SingleFeat(feat,1) for feat in dense_features]
    
#     print("SPARSE FEATURES: ", sparse_feat_list)
#     print("DENSE  FEATURES: ", dense_feat_list)

    # 3.generate input data for model
    train, test = df_train_trans.copy(), df_test_trans.copy()
    train_model_input = [train[feat.name].values for feat in sparse_feat_list+dense_feat_list]
    test_model_input  = [test[feat.name].values for feat in sparse_feat_list+dense_feat_list]
    
    if len(sequence_features) > 0:
        train_model_input = train_model_input+[genres_list_trn]
        test_model_input  = test_model_input+[genres_list_tst]
    
    
    # 4.Define Model,train,predict and evaluate
    dfm_params['feature_dim_dict'] = {"sparse": sparse_feat_list,"dense": dense_feat_list,"sequence": seq_feat_list}
    model = DeepFM(**dfm_params)
    model.compile(params['opt'], params['loss'], metrics=[params['loss']],)
    
    test_y  = test[target].values
    train_y = train[target].values
    

    print("-"*100)
    for i in range(n_iter):
        print("step: {}".format(i+1))
        history = model.fit(train_model_input, train_y,
                            batch_size=256, epochs=1, verbose=2, validation_split=0,)
        
        pred_trn_ans = model.predict(train_model_input, batch_size=500)
        pred_ans = model.predict(test_model_input, batch_size=500)
        
        mae[i]  = mean_absolute_error(test_y, pred_ans)
        rmse[i] = np.sqrt(mean_squared_error(test_y, pred_ans))
        mae_trn[i]  = mean_absolute_error(train_y, pred_trn_ans)
        rmse_trn[i] = np.sqrt(mean_squared_error(train_y, pred_trn_ans))
        mse[i] = mean_squared_error(test_y, pred_ans)
        mse_trn[i] = mean_squared_error(train_y, pred_trn_ans)
        
#         print("test MSE", round(mse[i], 4))
        print("-"*100)
        
    return mae, rmse, mse, mae_trn, rmse_trn, mse_trn
```

```python
def cross_valid(data, dfm_params, sparse_features, dense_features, sequence_features, params):
    n_iter = params['n_iter']
    mae_ttl,  mae_trn_ttl  = np.zeros(n_iter), np.zeros(n_iter)
    rmse_ttl, rmse_trn_ttl = np.zeros(n_iter), np.zeros(n_iter)
    mse_ttl,  mse_trn_ttl  = np.zeros(n_iter), np.zeros(n_iter)
    
    v_size = len(data)
    
    for i, d in enumerate(data):
        print("\nDataCase {}".format(i+1))
        df_train, df_test = d
        mae,rmse,mse, mae_trn,rmse_trn,mse_trn = valid_deepfm(df_train, df_test, dfm_params, 
                                                      sparse_features, dense_features, sequence_features, params)
        mae_ttl, mae_trn_ttl   = mae_ttl+np.array(mae), mae_trn_ttl+np.array(mae_trn)
        rmse_ttl, rmse_trn_ttl = rmse_ttl+np.array(rmse), rmse_trn_ttl+np.array(rmse_trn)
        mse_ttl, mse_trn_ttl   = mse_ttl+np.array(mse), mse_trn_ttl+np.array(mse_trn)
        
    return mae_ttl/v_size, mae_trn_ttl/v_size, rmse_ttl/v_size, rmse_trn_ttl/v_size, mse_ttl/v_size, mse_trn_ttl/v_size
```

```python
def test(sparse_features, dense_features, n_iter=50, h_size=[50, 50, 30, 10], emb_size=8, 
         keep_prob=1.0, final_act='linear', opt='adam', loss='mse', use_bn=False):
    dfm_params = dict(embedding_size=emb_size, hidden_size=h_size, 
                      l2_reg_linear=10, l2_reg_embedding=10, 
                      keep_prob=keep_prob, final_activation=final_act, use_bn=use_bn)
    ## final_activation: linear, sigmoid, tanh, 
    params = {'n_iter':n_iter, 'opt':opt, 'loss':loss}
    ## opt: adam, sgd, adadelta, adagrad, adamax, nadam, rmsprop
    ## loss: mse, mae, 
    mae,mae_trn,rmse,rmse_trn,mse,mse_trn = cross_valid(dataset, dfm_params, sparse_features, dense_features, params)

    plot_result(mae_trn,mae,rmse_trn,rmse,mse_trn,mse)
```

```python
class FeatureDictionary(object):
    def __init__(self, trainfile=None, testfile=None,
                 dfTrain=None, dfTest=None, numeric_cols=[], ignore_cols=[]):
        assert not ((trainfile is None) and (dfTrain is None)), "trainfile or dfTrain at least one is set"
        assert not ((trainfile is not None) and (dfTrain is not None)), "only one can be set"
        assert not ((testfile is None) and (dfTest is None)), "testfile or dfTest at least one is set"
        assert not ((testfile is not None) and (dfTest is not None)), "only one can be set"
        self.trainfile = trainfile
        self.testfile = testfile
        self.dfTrain = dfTrain
        self.dfTest = dfTest
        self.numeric_cols = numeric_cols
        self.ignore_cols = ignore_cols
        self.gen_feat_dict()

    def gen_feat_dict(self):
        if self.dfTrain is None:
            dfTrain = pd.read_csv(self.trainfile)
        else:
            dfTrain = self.dfTrain
        if self.dfTest is None:
            dfTest = pd.read_csv(self.testfile)
        else:
            dfTest = self.dfTest
        df = pd.concat([dfTrain, dfTest])
        self.feat_dict = {}
        tc = 0
        for col in df.columns:
            if col in self.ignore_cols:
                continue
            if col in self.numeric_cols:
                # map to a single index
                self.feat_dict[col] = tc
                tc += 1
            else:
                us = df[col].unique()
                self.feat_dict[col] = dict(zip(us, range(tc, len(us)+tc)))
                tc += len(us)
        self.feat_dim = tc
        
class DataParser(object):
    def __init__(self, feat_dict):
        self.feat_dict = feat_dict

    def parse(self, infile=None, df=None, has_label=False):
        assert not ((infile is None) and (df is None)), "infile or df at least one is set"
        assert not ((infile is not None) and (df is not None)), "only one can be set"
        if infile is None:
            dfi = df.copy()
        else:
            dfi = pd.read_csv(infile)
        if has_label:
            y = dfi["target"].values.tolist()
            dfi.drop(["id", "target"], axis=1, inplace=True)
        else:
            ids = dfi["id"].values.tolist()
            dfi.drop(["id"], axis=1, inplace=True)
        # dfi for feature index
        # dfv for feature value which can be either binary (1/0) or float (e.g., 10.24)
        dfv = dfi.copy()
        for col in dfi.columns:
            if col in self.feat_dict.ignore_cols:
                dfi.drop(col, axis=1, inplace=True)
                dfv.drop(col, axis=1, inplace=True)
                continue
            if col in self.feat_dict.numeric_cols:
                dfi[col] = self.feat_dict.feat_dict[col]
            else:
                dfi[col] = dfi[col].map(self.feat_dict.feat_dict[col])
                dfv[col] = 1.

        # list of list of feature indices of each sample in the dataset
        Xi = dfi.values.tolist()
        # list of list of feature values of each sample in the dataset
        Xv = dfv.values.tolist()
        if has_label:
            return Xi, Xv, y
        else:
            return Xi, Xv, ids
```
