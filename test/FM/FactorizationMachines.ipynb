````python
def split_IOdata(df):
    df_ = df.copy()

    ## y: ndarray, x: scipy_csc_matrix に変換
    x = sparse.csc_matrix(df_.drop(columns='rating').values)
    y = np.array(df_['rating'], dtype='float')
    
    return x, y
    
def make_valdata(val_case, test_size=20000, select_cols=False):
    df_train = pd.read_csv("../../dataset/ml-100k/valcase_{}/train_data_with_datetime.csv".format(val_case), names=['user_id','movie_id','rating','time_stamp'])
    df_test  = pd.read_csv("../../dataset/ml-100k/valcase_{}/test_data_with_datetime.csv".format(val_case),  names=['user_id','movie_id','rating','time_stamp'])
    
    df_all = pd.concat([df_train, df_test], axis=0)
    df_all = pd.merge(df_all, df_user, how='left', on='user_id')
    df_all = pd.merge(df_all, df_item, how='left', on='movie_id')
    if select_cols != False:
        df_all = df_all[select_cols]
    
    df_all[['user_id', 'movie_id']] = df_all[['user_id', 'movie_id']].astype(np.str)
    df_all = pd.get_dummies(df_all)
    
    df_train = df_all.iloc[:-test_size,:]
    df_test  = df_all.iloc[-test_size:,:]

    ## 特徴量, 教師ラベルを取得
    train_x, train_y = split_IOdata(df_train)
    test_x,  test_y  = split_IOdata(df_test)
    
    ## train_y → [-1, 1]の範囲に正規化
#     train_y = (train_y-1)/4*2-1
    
    return train_x, train_y, test_x, test_y
````

````python
def set_model(model_name, train_x, train_y, test_x, alpha=0.1, rank=100):
    ## SGD
    if model_name == 'sgd':
        fm_model = sgd.FMRegression(n_iter=1, init_stdev=0.1, rank=rank, random_state=123, l2_reg_w=alpha, 
                                    l2_reg_V=alpha, l2_reg=0, step_size=1)
    ## ALS
    elif model_name == 'als':
        fm_model = als.FMRegression(n_iter=0, init_stdev=0.1, rank=rank, random_state=123, l2_reg_w=alpha, 
                                    l2_reg_V=alpha, l2_reg=0)
        fm_model.fit(train_x, train_y)
    ## MCMC
    elif model_name == 'mcmc':
        fm_model = mcmc.FMRegression(n_iter=0, init_stdev=0.1, rank=rank, random_state=123)
        fm_model.fit_predict(train_x, train_y, test_x)
    
    return fm_model

def fit_and_predict(fm_model, train_x, train_y, test_x, model_name):
    ## SGD
    if model_name == 'sgd':
        fm_model.fit(train_x, train_y)
        return fm_model.predict(test_x)
    ## ALS
    elif model_name == 'als':
        fm_model.fit(train_x, train_y, n_more_iter=1)
        return fm_model.predict(test_x)
    ## MCMC
    elif model_name == 'mcmc':
        return fm_model.fit_predict(train_x, train_y, test_x, n_more_iter=1)
````

````python
def corss_val(algorism, data, alpha=10, rank=5, n_iter=100):
    mae_arr_cv, rmse_arr_cv, mse_arr_cv =  np.empty((5, n_iter)), np.empty((5, n_iter)), np.empty((5, n_iter))

    for j, d_case in enumerate(range(1,6)):
        train_x, train_y, test_x, test_y = data[j]
        fm_model = set_model(algorism, train_x, train_y, test_x, alpha=alpha, rank=rank)
        mae, rmse, mse = [0 for _ in range(n_iter)], [0 for _ in range(n_iter)], [0 for _ in range(n_iter)]

        for k in range(n_iter):
#             print("iter_{}".format(k))
            y_pred = fit_and_predict(fm_model, train_x, train_y, test_x, model_name=algorism)

            mae[k]  = mean_absolute_error(test_y, y_pred)
            rmse[k] = np.sqrt(mean_squared_error(test_y, y_pred))
            mse[k]  = mean_squared_error(test_y, y_pred)

        mae_arr_cv[j]   = np.array(mae)
        rmse_arr_cv[j]  = np.array(rmse)
        mse_arr_cv[j]   = np.array(mse)
        print("valcase_{} comp".format(d_case))
        
    return mae_arr_cv.mean(axis=0), rmse_arr_cv.mean(axis=0), mse_arr_cv.mean(axis=0)

def fm_eval(algorism, h_params, h_params_name, n_iter=100):
    mae_arr, rmse_arr, mse_arr =  np.empty((len(h_params), n_iter)),np.empty((len(h_params), n_iter)),np.empty((len(h_params), n_iter))
    
    if h_params_name == 'alpha':

        for i, h_param in enumerate(h_params):
#             print("{0} {1}".format(h_params_name, h_param))    
            
            mae_arr[i], rmse_arr[i], mse_arr[i] = corss_val(algorism, dataset, alpha=h_param, n_iter=n_iter)
#             print("{0} {1} comp".format(h_params_name, h_param))
#             print("-"*100)

        return mae_arr, rmse_arr, mse_arr
    
    elif h_params_name == 'rank':
        for i, h_param in enumerate(h_params):
#             print("{0} {1}".format(h_params_name, h_param))    

            mae_arr[i], rmse_arr[i], mse_arr[i] = corss_val(algorism, dataset, rank=h_param, n_iter=n_iter)
#             print("{0} {1} comp".format(h_params_name, h_param))
#             print("-"*100)

        return mae_arr, rmse_arr, mse_arr
````

````python
def plot_result(mae, rmse, mse, h_params, h_params_name, colors, steps):
    mae_list  = np.array([v[steps-10:steps].mean() for v in mae])
    rmse_list = np.array([v[steps-10:steps].mean() for v in rmse])
    mse_list  = np.array([v[steps-10:steps].mean() for v in mse])
    mae_max, mae_min   = mae_list.max(), mae_list.min()
    rmse_max, rmse_min = rmse_list.max(), rmse_list.min()
    mse_max, mse_min   = mse_list.max(), mse_list.min()
    df_result = pd.DataFrame([mae_list, rmse_list, mse_list], index=['MAE','RMSE','MSE'], columns=['{0}_{1}'.format(h_params_name, v) for v in h_params])

    ## MAE
    plt.figure(figsize=(10,6))
    for idx, (v, color) in enumerate(zip(h_params, colors[:len(h_params)])):
        plt.plot(mae[idx], color=color, label='{0}{1}'.format(h_params_name, v))
    plt.title("Mean Absolute Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("MAE", fontsize=12)
#     plt.legend(loc='upper right', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()
    
    plt.figure(figsize=(2.3*len(mae_list),6))
    plt.bar([i for i,_ in enumerate(mae_list)], mae_list, tick_label=list(h_params), align="center", color='#dc143c', width=0.5)
#     plt.plot(h_params, mae_list, color='#dc143c', marker="o")
    plt.title("Mean Absolute Error / {0} / step_{1}".format(h_params_name[:-1], steps), fontsize=12)
    plt.xlabel(h_params_name[:-1], fontsize=12)
    plt.ylabel("MAE", fontsize=12)
    plt.ylim([mae_min-(mae_max-mae_min)/10,mae_max+(mae_max-mae_min)/10])
    plt.grid(True)
    plt.show()
    
    
    ## RMSE
    plt.figure(figsize=(10,6))
    for idx, (v, color) in enumerate(zip(h_params, colors[:len(h_params)])):
        plt.plot(rmse[idx], color=color, label='{0}{1}'.format(h_params_name, v))
    plt.title("Root Mean Square Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("RMSE", fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()

#     plt.figure(figsize=(16,6))
    plt.figure(figsize=(2.3*len(rmse_list),6))
    plt.bar([i for i,_ in enumerate(rmse_list)], rmse_list, tick_label=list(h_params), align="center", color='#dc143c', width=0.5)
#     plt.plot(h_params, rmse_list, color='#dc143c', marker="o")
    plt.title("Root Mean Square Error / {0} / step_{1}".format(h_params_name[:-1], steps), fontsize=12)
    plt.xlabel(h_params_name[:-1], fontsize=12)
    plt.ylabel("RMSE", fontsize=12)
    plt.ylim([rmse_min-(rmse_max-rmse_min)/10,rmse_max+(rmse_max-rmse_min)/10])
    plt.grid(True)
    plt.show()
    
    
    ## MSE
    plt.figure(figsize=(10,6))
    for idx, (v, color) in enumerate(zip(h_params, colors[:len(h_params)])):
        plt.plot(mse[idx], color=color, label='{0}{1}'.format(h_params_name, v))
    plt.title("Mean Square Error", fontsize=12)
    plt.xlabel("steps", fontsize=12)
    plt.ylabel("MSE", fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
    plt.grid(True)
    plt.show()

#     plt.figure(figsize=(16,6))
    plt.figure(figsize=(2.3*len(mse_list),6))
    plt.bar([i for i,_ in enumerate(mse_list)], mse_list, tick_label=list(h_params), align="center", color='#dc143c', width=0.5)
#     plt.plot(h_params, rmse_list, color='#dc143c', marker="o")
    plt.title("Mean Square Error / {0} / step_{1}".format(h_params_name[:-1], steps), fontsize=12)
    plt.xlabel(h_params_name[:-1], fontsize=12)
    plt.ylabel("MSE", fontsize=12)
    plt.ylim([mse_min-(mse_max-mse_min)/10,mse_max+(mse_max-mse_min)/10])
    plt.grid(True)
    plt.show()
    
    

    return df_result
#     df_result = pd.DataFrame([mae[0], rmse[0], mse[0]], index=['MAE','RMSE','MSE'])
#     return pd.DataFrame(df_result.loc[:, df_result.loc['RMSE'].idxmin()])
````
